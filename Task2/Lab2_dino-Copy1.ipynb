```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('abc.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
template = image[675:725, 70:120].copy()
#template = image[675:725, 470:520].copy()
plt.figure(figsize=(1, 1))
plt.imshow(template)
plt.show()
```

```python
match = cv2.matchTemplate(image, template, cv2.TM_CCORR_NORMED)
plt.imshow(match)
plt.colorbar()
plt.show()
```

```python
np.where(match == match.max())
```
