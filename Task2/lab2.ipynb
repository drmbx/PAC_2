```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import torchvision
```

```python
sources = ["A_B_C.jpg", "Ab_C.jpg", "abc.png", "abc.jpg"]
images = []
for source in sources:
    images.append(cv2.cvtColor(cv2.imread(source), cv2.COLOR_BGR2RGB))
plt.imshow(images[1])
plt.show()
```

```python
model = torchvision.models.resnet50(weights=torchvision.models.ResNet50_Weights.IMAGENET1K_V1)
```

```python
print(np.shape(images[1]))
template = images[1][408:484, 210:340]
print(np.shape(template))
plt.imshow(template)
plt.show()
```

```python
match = cv2.matchTemplate(images[1], template, cv2.TM_CCORR_NORMED)
plt.imshow(match)
plt.colorbar()
plt.show()
```
